```python
import pandas as pd
import numpy as ny
```

```python
sharkTankData_csv = "../data/raw/shark_tank.csv"
sharkTankDeals_csv = "../data/raw/Sharktankpitchesdeals.csv"
```

```python
sharkTankData_df = pd.read_csv(sharkTankData_csv)
sharkTankDeals_df = pd.read_csv(sharkTankDeals_csv)
```

```python
sharkTankData_df = sharkTankData_df.rename(columns= {"deal": "Deal", "description": "Description", "episode": "Episode", "category":"Category", "entrepreneurs": "Entrepreneurs",  "location" : "Location",  "website" : "Website",  "askedFor": "Amount Asked For", "exchangeForStake": "Exchange For Stake",  "valuation": "Valuation",  "season": "Season",   "shark1": "Shark 1", "shark2": "Shark 2",  "shark3": "Shark 3",  "shark4": "Shark 4",  "shark5": "Shark 5",  "title" : "Title",  "episode-season" : "Episode-Season"})
```

```python
sharkTankData_df.astype({'Valuation' : 'float'})
```

```python
sharkTankDeals_df = sharkTankDeals_df.rename(columns={ 'Pitched_Business_Identifier': 'Title', "Season_Epi_code": "Episode Code",  "Pitched_Business_Desc": "Pitched Business Description", "Deal_Status": "Deal Status",  "Deal_Shark": "Shark(s) That Made a Deal"})
  
```

```python
sharkTankData_df['Amount Asked For'] = sharkTankData_df['Amount Asked For'].map('${:,.2f}'.format) 
sharkTankData_df['Valuation'] = sharkTankData_df['Valuation'].map('${:,.2f}'.format) 
sharkTankData_df['Exchange For Stake'] = sharkTankData_df['Exchange For Stake'].map('{:.2f}%'.format) 
```

```python
sharkTankDeals_df.astype({'Deal Status' : 'str'}).dtypes
```

```python
sharkTankDeals_df["Deal Status"]= sharkTankDeals_df["Deal Status"].replace(1, "Deal Made") 

sharkTankDeals_df["Deal Status"]= sharkTankDeals_df["Deal Status"].replace(0, "No Deal Made") 
```

```python
mergedSharkTank_df = pd.merge(sharkTankData_df, sharkTankDeals_df,
                                 how='outer', on='Title')
mergedSharkTank_df.head()
```
